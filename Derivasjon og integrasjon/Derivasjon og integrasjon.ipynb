```python
"7.3"
from pylab import*
antall = []
tid = []
endringen = []

def N(t):
    utvikling = 0.2 * 2.718**t + 90
    return utvikling

def dirivert(N, t, dt):
    Nder = (N(t + dt) - N(t))/dt
    return Nder
    
for i in range(10):
    tid.append(i)
    antall.append(N(i))
    endringen.append(dirivert(N, i, 10**-8))
    

plot(tid, antall)
plot(tid, endringen)
legend(["Planter","endringen i antall planter"])
show()
```

```python
"7.13"
def f(x):
    return 2*x + 1

def integralet(f, a, b, N):
    A = 0
    dx = (b - a)/N
    for i in range(N):
        A = A + f(a + i*dx)*dx
    return A

print(integralet(f, 1, 4, 1000))
```

```python
"7.12"
from pylab import *

data = loadtxt("posisjon.txt", skiprows = 1)
tid = data[:,0]
posisjon = data[:,1]
fart = []
aks = []

for i in range(len(tid)-1):
    dy = posisjon[i+1]-posisjon[i]
    dx = tid[i+1]-tid[i]
    der = dy/dx
    fart.append(der)

for i in range(len(tid)-2):
    dy = fart[i+1]-fart[i]
    dx = tid[i+1]-tid[i]
    der = dy/dx
    aks.append(der)
    
plot(tid, posisjon)
plot(tid[:-1], fart)
plot(tid[:-2], aks)

legend(["Posisjon","Fart","Akselerasjon"])
xlabel("Tid")
```
