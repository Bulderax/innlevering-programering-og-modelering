# Modelleringsoppgave I: Smittespredning
## oppgave 1
Modelen passer bra i et kort tidsrom etter en sykdom har begynt å spre seg. Da sprer sykdomen seg i et raskt tempo noe modelen kan forutsi.

begrensningene er at en kan bli smitted flere ganger. Dette kan virke urelistisk i mange tilfeller. 
endren man antal smitede til et støre tall, vil grafen starte med flere smitede. det vil ikke ha noen serlig effekt på denne modelen.
det er viktig at smiteraten har en enhet: uker, fordi ellers vile vi ikke vist hvor fort smiteraten er.



```python
from pylab import *
u = 48     # slut tid uker
p = 157759 # indevider
k = 0.2    #smitte rate
t = 0      # antal smitede
smitte = []
tid = []

for i in range(u):
    t = t + p*k
    
    
    smitte.append(t)
    tid.append(i)
    
plot(tid,smitte)
show()
```

## oppgave 2

i den nye modelen tar vi hensyn til at etter å ha vært smitet vil man bli imun.
grafen stiger ekspotensielt når flere og fler blir smittet.
Når du variere antall smitede og a (smitte raten) vil atall smitede stige enda fortere. grafen vil ikke gå like høyt som den i oppgave 1 fordi når folk blir imune går sykdomen tom for folk den kan smitte.
Det kan være en fin model å bruke etter smiten har spred seg en stund.

```python
from pylab import *
slut_tid = 48 #uker
start_tid = 0
p = 157759 # indevider
a = 0.2    #smitte rate
t = 3      # antal smitede
smitte = []
tid = []
mottakelige = []

for i in range(start_tid, slut_tid + 1):
    t = t + t *a
    smitte.append(t)
    tid.append(i)
    s = p - t
    s = mottakelige.append(s)
    
print(s)
plot(tid,smitte, label="syke")
plot(tid,mottakelige, label="mottakelige")
show()
```

## oppgave 3


```python
from pylab import*

# Startbetingelser
N = 157759 # Antall individer
a = 4/N    # Smitterate
b = 3.5    # Bedringsrate
tid = 48   # Tid

I = 3      # Smittede
S = N - I  # Mottakelige
R = 0      # Restituert/friske

syke = [I]
mulige = [S]
friske = [R]
t = [0]

for i in range(1,tid):
    S = S - a*S*I
    I = I + a*S*I - b*I
    R = R + b*I
    syke.append(I)
    mulige.append(S)
    friske.append(R)
    t.append(i)
    
plot(t,syke,label='Syke')
plot(t,mulige,label='Mottakelige')
plot(t,friske,label='Friske')
legend()
show()
```
